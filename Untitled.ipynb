### False Nulls
According to the data dictionary, 'NA' can be used to indicate that a particular property attribute does not exist, amidst an otherwise ordinal list of ranking that attribute.  For instance, 'Rating of basement finished area' is on a scale of Unfinished up to Good Living Quarters, with NA indicating that there is no basement.  In this scenario, NA can = 0, Unfinished = 1, and Good Living Quarters = max.

NA is its own level of ordinal rank in the case of 11 variables.  6 share the same 6-level scale (NA, Po, Fa, TA, Gd, Ex).  The Pool category has all of these levels except for 'Po', so I will include it with these 6 variables as I numify the rankings (Pool rankings will then consist of NA=0, Fa=2, ... , Ex=6).  Apparently in Ames, any pool is already considered 'Fair', but as a water sport athlete I'd argue there are DEFINITELY 'Poor' pools and Ames county is ignoring a valuable metric.
